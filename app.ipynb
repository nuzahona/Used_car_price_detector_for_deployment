```python
def convert_kmkg_to_kmpl(mileage):
    if isinstance(mileage, str) and 'km/kg' in mileage:
        value = float(mileage.replace(' km/kg', '').strip())
        return value * 1.39  # Conversion to kmpl
    return mileage

lily['mileage'] = lily['mileage'].apply(convert_kmkg_to_kmpl)
```

```python
median_price = lily['selling_price'].median()
print(median_price)
```

```python
lily['price_category'] = lily['selling_price'].apply(lambda x: 0 if x <= median_price else 1)
```

```python
lily
```

```python
X = lily.drop(columns=['selling_price', 'price_category'])  # Features (excluding target and price category)
Y = lily['price_category']
```

```python
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2, random_state=2)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)  # Fit and transform on training data
X_test_scaled = scaler.transform(X_test)
```

```python
#train model - support-vector machine model
SVM_model = svm.SVC(kernel = 'linear')
```

```python
#train data
SVM_model.fit(X_train_scaled, Y_train)
```

```python
#accuracy score
X_train_prediction = SVM_model.predict(X_train_scaled)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
print("Data accuracy:" ,training_data_accuracy)

```

```python
#test data accuracy
X_test_prediction = SVM_model.predict(X_test_scaled)
SVM_accuracy = accuracy_score(X_test_prediction, Y_test)

print("SVC accuracy:" ,SVM_accuracy)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_model = LogisticRegression(class_weight='balanced', max_iter=1000)
```

```python
log_model.fit(X_train_scaled, Y_train)
```

```python
X_train_prediction = log_model.predict(X_train_scaled)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)

print("Data accuracy:" ,training_data_accuracy)
```

```python
X_test_prediction = log_model.predict(X_test_scaled)
LogisticRegression_accuracy = accuracy_score(X_test_prediction, Y_test)

print("LogisticRegression accuracy:" ,LogisticRegression_accuracy)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn_model = KNeighborsClassifier(n_neighbors=14)
```

```python
knn_model.fit(X_train_scaled, Y_train)
```

```python
X_train_prediction = knn_model.predict(X_train_scaled)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)

print("Data accuracy:" ,training_data_accuracy)
```

```python
X_test_prediction = knn_model.predict(X_test_scaled)
KNeighbors_accuracy = accuracy_score(X_test_prediction, Y_test)

print("KNeighbors accuracy:" ,KNeighbors_accuracy)
```

```python
def getBestNeighbor(x):
    best_k=4
    best_acc = 0
    k = 4
    for i in range(100,x):
        knn_model = KNeighborsClassifier(n_neighbors=i)
        knn_model.fit(X_train_scaled, Y_train)
        X_test_prediction = knn_model.predict(X_test_scaled)
        test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
        if test_data_accuracy > best_acc:
            best_acc = test_data_accuracy
            best_k = i
    return best_k


print(getBestNeighbor(200))
```

```python
import matplotlib.pyplot as plt

# Assuming you have the accuracy scores for CNN, SVM, and Naive Bayes models


# Accuracy scores
accuracies = [ SVM_accuracy, LogisticRegression_accuracy, KNeighbors_accuracy]
models = [ 'SVC_Classifier', 'Logistic Regression', 'KNeighbors Classifier']


plt.figure(figsize=(8, 6))
plt.bar(models, accuracies, color=['blue', 'orange', 'green', 'red'])
plt.title('Accuracy Comparison of Different Models')
plt.xlabel('Models')
plt.ylabel('Accuracy')
plt.ylim(0, 1)  # Setting the y-axis limit from 0 to 1 for accuracy range
plt.show()
```

```python
from sklearn.metrics import precision_score, recall_score

# Support Vector Machine (SVM) Precision and Recall
svm_predictions = SVM_model.predict(X_test_scaled)
svm_precision = precision_score(Y_test, svm_predictions)
svm_recall = recall_score(Y_test, svm_predictions)

print("SVM Precision:", svm_precision)
print("SVM Recall:", svm_recall)

# Logistic Regression Precision and Recall
log_predictions = log_model.predict(X_test_scaled)
log_precision = precision_score(Y_test, log_predictions)
log_recall = recall_score(Y_test, log_predictions)

print("Logistic Regression Precision:", log_precision)
print("Logistic Regression Recall:", log_recall)

# K-Nearest Neighbors (KNN) Precision and Recall
knn_predictions = knn_model.predict(X_test_scaled)
knn_precision = precision_score(Y_test, knn_predictions)
knn_recall = recall_score(Y_test, knn_predictions)

print("KNN Precision:", knn_precision)
print("KNN Recall:", knn_recall)

# Compare Precision and Recall
print("\nModel Comparison:")
print(f"SVM -> Precision: {svm_precision:.2f}, Recall: {svm_recall:.2f}")
print(f"Logistic Regression -> Precision: {log_precision:.2f}, Recall: {log_recall:.2f}")
print(f"KNN -> Precision: {knn_precision:.2f}, Recall: {knn_recall:.2f}")
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# Support Vector Machine (SVM) Confusion Matrix
svm_predictions = SVM_model.predict(X_test_scaled)
svm_cm = confusion_matrix(Y_test, svm_predictions)

print("SVM Confusion Matrix:")
print(svm_cm)
ConfusionMatrixDisplay(svm_cm).plot(cmap='Blues')
plt.title("SVM Confusion Matrix")
plt.show()

# Logistic Regression Confusion Matrix
log_predictions = log_model.predict(X_test_scaled)
log_cm = confusion_matrix(Y_test, log_predictions)

print("Logistic Regression Confusion Matrix:")
print(log_cm)
ConfusionMatrixDisplay(log_cm).plot(cmap='Oranges')
plt.title("Logistic Regression Confusion Matrix")
plt.show()

# K-Nearest Neighbors (KNN) Confusion Matrix
knn_predictions = knn_model.predict(X_test_scaled)
knn_cm = confusion_matrix(Y_test, knn_predictions)

print("KNN Confusion Matrix:")
print(knn_cm)
ConfusionMatrixDisplay(knn_cm).plot(cmap='Greens')
plt.title("KNN Confusion Matrix")
plt.show()
```
